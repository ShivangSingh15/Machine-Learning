```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df_train= pd.read_csv("C:\\Users\\shiva\\Downloads\\train\\train.csv")
df_train
```

```python
df_test= pd.read_csv("C:\\Users\\shiva\\Downloads\\test\\test.csv")
df_test
```

```python
y_train= df_train['y'].values
id_test= df_test['ID'].values
```

```python
usable_columns= list(set(df_train.columns) - set(['ID', 'y']))
x_train= df_train[usable_columns]
x_test= df_test[usable_columns]
```

```python
# Check for null and unique values for test and train sets.

def check_missing_values(df):
    if df.isnull().any().any():
        print("There are missing values in the dataframe")
    else:
        print("There are no missing values in the dataframe")
check_missing_values(x_train)
check_missing_values(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
# Apply label encoder.

x_train[usable_columns]= x_train[usable_columns].apply(le.fit_transform)
x_train
```

```python
# Apply label encoder.

x_test[usable_columns]= x_test[usable_columns].apply(le.fit_transform)
x_test
```

```python
# If for any column(s), the variance is equal to zero, then you need to remove those variable(s).

for column in usable_columns:
    if x_train[column].var() == 0:
        x_train.drop(column, axis= 1, inplace= True)
        x_test.drop(column, axis= 1, inplace= True)

x_train
```

```python
x_train.info()
```

```python
from sklearn.decomposition import PCA
```

```python
# Perform dimensionality reduction.

pca= PCA(n_components= 10)
pca_train= pca.fit_transform(x_train)
pca_test= pca.transform(x_test)
```

```python
import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_valid, y_train, y_valid= train_test_split(pca_train, y_train, test_size= 0.2, random_state= 1)
```

```python
xgb_train= xgb.DMatrix(x_train, label= y_train)
xgb_valid= xgb.DMatrix(x_valid, label= y_valid)
xgb_test= xgb.DMatrix(pca_test)
```

```python
# training our dataset using XGBOOST 

params= {}
params['objective'] = 'reg:linear'
params['eta'] = 0.02
params['max_depth'] = 4

def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)

watchlist = [(xgb_train, 'train'), (xgb_valid, 'valid')]

clf = xgb.train(params, xgb_train, 
                1000, watchlist, early_stopping_rounds=50, 
                feval=xgb_r2_score, maximize=True, verbose_eval=10)
```

```python
# Predict your test_df values using XGBoost.

p_test = clf.predict(xgb_test)

sub = pd.DataFrame()
sub['ID'] = id_test
sub['y'] = p_test
sub.to_csv('xgb.csv', index=False)

sub.head()
```
